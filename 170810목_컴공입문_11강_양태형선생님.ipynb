```python
#어제내준 숙제풀이
#pseudo_constructor
#initialize object members
def person_init(name, money):
    obj = {'name' : name, 'money' : money} #member
    obj['give_money'] = Person[1]
    obj['get_money'] = Person[2] #위 2줄이 method
    obj['show'] = Person[3]
    return obj

#instance method
def give_money(self, other, money): #첫번째 인자로 self를 넣어주기 
                #때문에 아래줄에 self를 넣어주어야 함.
    self['money'] -= money 
    other['get_money'](other, money)
    
#instance method
def get_money(self, money):
    self['money'] += money

#instance method
def show(self):
    print('{} : {}'.format(self['name'], self['money']))
    
#class
#container of instance methods
Person = person_init, give_money, get_money, show

#object creation
g = Person[0]('greg', 5000)
j = Person[0]('john', 2000)

g['show'](g)
j['show'](j)
greg : 5000
john : 2000

#message passing
g['give_money'](g, j, 2000)

#does it work?
g['show'](g)
j['show'](j)
greg : 3000
john : 4000

```
